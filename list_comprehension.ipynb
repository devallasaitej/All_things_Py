Comparing List comprehension with For loop

```python
import time

numbers = list(range(1, 1000000))

# for-loop
start_time = time.time()
doubled_numbers_loop = []
for number in numbers:
    doubled_numbers_loop.append(number * 2)
end_time = time.time()
print(f"For loop : {end_time - start_time} seconds")

# list comprehension
start_time = time.time()
doubled_numbers_comprehension = [number * 2 for number in numbers]
end_time = time.time()
print(f"List comprehension : {end_time - start_time} seconds")
```

```python
import time
import matplotlib.pyplot as plt

def measure_performance():
    sizes = [1000, 5000, 10000, 50000, 100000, 500000] 
    loop_times = [] 
    comprehension_times = [] 

    for size in sizes:
        numbers = list(range(1, size+1))

        # for-loop
        start_time = time.time()
        doubled_numbers_loop = []
        for number in numbers:
            doubled_numbers_loop.append(number * 2)
        end_time = time.time()
        loop_times.append(end_time - start_time)

        # list comprehension
        start_time = time.time()
        doubled_numbers_comprehension = [number * 2 for number in numbers]
        end_time = time.time()
        comprehension_times.append(end_time - start_time)

    return sizes, loop_times, comprehension_times
```

```python
sizes, loop_times, comprehension_times = measure_performance()

plt.figure(figsize=(10, 6))
plt.plot(sizes, loop_times, label='For-Loop', marker='o')
plt.plot(sizes, comprehension_times, label='List Comprehension', marker='o')

plt.xlabel('Data Size')
plt.ylabel('Execution Time (seconds)')
plt.title('Performance Comparison: For-Loop vs List Comprehension')
plt.legend()
plt.grid(True)
plt.show()
```
